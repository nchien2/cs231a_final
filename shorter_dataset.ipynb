```python
from google.colab import drive
import os
import csv

drive.mount('/content/drive', force_remount=True)
```

```python
old_csv_dir = 'drive/MyDrive/CS231A/Project/asllvd_signs_2022_11_29.csv'
new_csv_dir = 'drive/MyDrive/CS231A/Project/asllvd_shorter_data.csv'
json_dir = 'drive/MyDrive/CS231A/Project/ASLLVD/output_json/'

least_num = 6 # least number of instances to be kept in the dataset
label_dict = dict() # dict that mapts old label to number of occurance
label_index = 1 # the index of label header in the original file

with open(old_csv_dir, 'r', newline='') as old_csv_file:
    with open(new_csv_dir, 'w', newline='') as new_csv_file:
        csv_reader = csv.reader(old_csv_file, delimiter=',')
        csv_writer = csv.writer(new_csv_file, delimiter=',')

        next(old_csv_file) # skip header row
        i = 0
        for row in csv_reader:
            i += 1
            if not os.path.exists(json_dir + str(i)):
                continue
            label = row[label_index]
            if label in label_dict:
                label_dict[label] += 1
            else:
                label_dict[label] = 1
        
        old_csv_file.seek(0) # back to first line
        # add two columns
        label_num = 0
        label_num_dict = dict() # dict that maps old label to new label (int)
        i = -1
        for row in csv_reader:
            i += 1
            if i == 0:
                header = ['json index', 'label index'] + row
                csv_writer.writerow(header)
            else:
                if not os.path.exists(json_dir + str(i)):
                    continue
                label = row[label_index]
                if label_dict[label] >= least_num:
                    if label not in label_num_dict:
                        label_num_dict[label] = label_num
                        label_num += 1
                    new_label = label_num_dict[label]
                    csv_writer.writerow([i, new_label] + row)
```
